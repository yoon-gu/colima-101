# 실험: uv로 CUDA 툴킷(nvcc) 12.4 맞추기

Colab 시스템 nvcc는 12.8. NVIDIA가 `nvidia-cuda-nvcc-cu12`를 pip로 공개하므로
uv venv에 **nvcc 12.4**를 따로 설치해 시스템과 무관하게 12.4를 쓸 수 있는지 확인한다.

```python
!pip install -q uv
!uv venv --python 3.11.9 /content/py311
!echo '== 시스템 nvcc =='; nvcc --version 2>/dev/null | tail -2 || echo '(시스템 nvcc 없음)'
```

```python
# venv에 CUDA 12.4 nvcc + 런타임 + 헤더(cccl) 설치
!uv pip install --python /content/py311/bin/python \
    "nvidia-cuda-nvcc-cu12>=12.4,<12.5" \
    "nvidia-cuda-runtime-cu12>=12.4,<12.5" \
    "nvidia-cuda-cccl-cu12>=12.4,<12.5"
print('설치 완료')
```

```python
!echo '== cuda_nvcc/bin 내용 =='; ls -la /content/py311/lib/python3.11/site-packages/nvidia/cuda_nvcc/bin
!echo '== nvcc --version =='; /content/py311/lib/python3.11/site-packages/nvidia/cuda_nvcc/bin/nvcc --version | tail -4 2>&1 || echo '(nvcc 실행 불가)'
```

## 결론

- `nvidia-cuda-nvcc-cu12==12.4.131`를 uv로 설치하면 `cuda_nvcc/bin/` 에 **`ptxas`만** 들어있고
  **`nvcc` 드라이버 본체는 없습니다**(실측). 즉 uv/pip로는 동작하는 nvcc 12.4를 손쉽게 못 얻습니다.
- **진짜 nvcc(컴파일러)** 가 필요하면 conda(`cuda-nvcc=12.4` / `cuda-toolkit=12.4`, nvidia 채널)나
  공식 설치 프로그램을 써야 합니다.
- ✅ 단, **CUDA 런타임 12.4 + cuDNN 9.1은 torch의 cu124 휠이 이미 설치**합니다
  (`nvidia-cuda-runtime-cu12 12.4`, `nvidia-cudnn-cu12 9.1`). **코드 실행/학습 목적이면
  이미 Pod과 동일**하므로 별도로 toolkit을 맞출 필요가 없습니다. nvcc가 필요한 경우는
  "CUDA 커널을 소스에서 직접 컴파일"할 때뿐입니다.
